```python
import cv2
import os

# Set the folder path to save images
save_dir = 'dataset/train/you'
os.makedirs(save_dir, exist_ok=True)

# Open webcam
cap = cv2.VideoCapture(0)

count = 0
print("Press 's' to save image, 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Display the video frame
    cv2.imshow("Press 's' to save - Webcam", frame)

    key = cv2.waitKey(1)

    if key == ord('s'):
        filename = os.path.join(save_dir, f"you_{count}.jpg")
        cv2.imwrite(filename, frame)
        print(f"✅ Saved: {filename}")
        count += 1

        if count == 50:
            print("✅ Done! Collected 50 images.")
            break

    elif key == ord('q'):
        print("❌ Quit without finishing.")
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import os

os.makedirs("dataset/train/you", exist_ok=True)        # Already done
os.makedirs("dataset/train/not_you", exist_ok=True)     # For images of other people
```

```python
import shutil
import os

source_folder = "dataset/train/not_you"
existing_images = os.listdir(source_folder)
current_count = len(existing_images)

# Duplicate to reach at least 50
target_count = 50
i = current_count
while len(os.listdir(source_folder)) < target_count:
    for img in existing_images:
        new_name = f"copy_{i}_{img}"
        shutil.copy(os.path.join(source_folder, img), os.path.join(source_folder, new_name))
        i += 1
        if len(os.listdir(source_folder)) >= target_count:
            break

print("✅ Now have", len(os.listdir(source_folder)), "'not you' images.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = 128
BATCH_SIZE = 8

# Load and rescale the images
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 80% for training, 20% for validation
)

# Training data
train_data = datagen.flow_from_directory(
    "dataset/train",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary",
    subset="training"
)

# Validation data
val_data = datagen.flow_from_directory(
    "dataset/train",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary",
    subset="validation"
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')  # Binary classification
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
import matplotlib.pyplot as plt

# Accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)[0][0]
    if prediction > 0.5:
        print("❌ Not You")
    else:
        print("✅ It's You!")

# Example test
predict_image("dataset/train/you/you_0.jpg")
```

```python
from tensorflow.keras.models import load_model
import numpy as np
import cv2

# Load your trained model
model = load_model("my_face_recognition_model.h5")
IMG_SIZE = 128
```

```python
def preprocess_frame(frame):
    # Resize and normalize
    img = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0
    return np.expand_dims(img, axis=0)

# Start webcam
cap = cv2.VideoCapture(0)
print("📸 Press 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Flip the image for mirror view
    flipped = cv2.flip(frame, 1)

    # Prepare the image
    input_img = preprocess_frame(flipped)

    # Predict
    prediction = model.predict(input_img)[0][0]

    # Label
    label = "It's YOU ✅" if prediction < 0.5 else "Not YOU ❌"
    color = (0, 255, 0) if prediction < 0.5 else (0, 0, 255)

    # Draw label
    cv2.putText(flipped, label, (30, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.0, color, 2)
    cv2.imshow("Real-Time Face Recognition", flipped)

    # Break on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
